```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data = df.sample(100000,random_state=42)

x = data.drop(columns=["timestamp","Target"])
y = data["Target"]

x_scaled = scaler.fit_transform(x)

x = pd.DataFrame(x_scaled, columns=x.columns)

x.drop(columns=["year",'reactive_power','active_power_calculated_by_converter','active_power_raw','grid_power10min_average'], inplace=True)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

le = LinearRegression()
le.fit(x_train, y_train)
y_pred = le.predict(x_test)
print("R^2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
#  Random Forest
from sklearn.ensemble import RandomForestRegressor
re = RandomForestRegressor(n_estimators=50,
    max_depth=10,
    n_jobs=-1,  # all cpu cores
    random_state=42)
re.fit(x_train, y_train)

y_pred = re.predict(x_test)
print("R^2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
# svr
from sklearn.svm import SVR
se = SVR()
se.fit(x_train, y_train)
y_pred = se.predict(x_test)
print("R^2 Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
#  vif
from sklearn.preprocessing import StandardScaler
df1 = df.drop(columns=['year','month','day','hour', 'timestamp'])
x = df1.drop(columns=['Target'])
y = df1['Target']

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x = pd.DataFrame(x_scaled, columns=x.columns)

def calculate_vif(data):
    while True:
        vif = pd.DataFrame()
        vif['Features'] = data.columns
        vif['VIF'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
        max_vif = vif['VIF'].max()
        if max_vif > 5:
            max_vif_feature = vif.loc[vif['VIF'].idxmax(), 'Features']
            print(f"Dropping '{max_vif_feature}' with VIF: {max_vif}")
            data = data.drop(columns=[max_vif_feature])
        else:
            break
    return data

x = calculate_vif(x)
```
